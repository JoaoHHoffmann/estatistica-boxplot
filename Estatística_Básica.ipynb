```python
import pandas as pd
dado_hist = {
    'Empresa 1' : [4, 4, 9, 6, 5],
    'Empresa 2' : [9, 5, 6, 4, 15],
    'Empresa 3' : [6, 4, 2, 2, 25],
    'Empresa 4' : [4, 2, 35, 4, 35],
    'Empresa 5' : [2, 4, 9, 9, 45]
}
df = pd.DataFrame(dado_hist)
df
```

## Estatística Básica

```python
variancia = df.loc[:, 'Empresa 1'].var(ddof=0)
desviop = df.loc[:, 'Empresa 1'].std(ddof=0)
media = df.loc[:, 'Empresa 1'].mean()
mediana = df.loc[:, 'Empresa 1'].median()
moda = df.loc[:, 'Empresa 1'].mode()
max = df.loc[:, 'Empresa 1'].max()
min = df.loc[:, 'Empresa 1'].min()
print(f"Variancia da empresa 1: {variancia}")
print(f"Desvio Padrão da empresa 1: {desviop:.2f}")
print(f"Média da empresa 1: {media:.2f}")
print(f"Mediana da empresa 1: {mediana}")
print(f"Moda da empresa 1: {moda}")
print(f"Máximo da empresa 1: {max}")
print(f"Mínimo da empresa 1: {min}")
```

```python
variancia = df.loc[:, 'Empresa 2'].var(ddof=0)
desviop = df.loc[:, 'Empresa 2'].std(ddof=0)
media = df.loc[:, 'Empresa 2'].mean()
mediana = df.loc[:, 'Empresa 2'].median()
moda = df.loc[:, 'Empresa 2'].mode()
max = df.loc[:, 'Empresa 2'].max()
min = df.loc[:, 'Empresa 2'].min()
print(f"Variancia da empresa 2: {variancia:.2f}")
print(f"Desvio Padrão da empresa 2: {desviop:.2f}")
print(f"Média da empresa 2: {media:.2f}")
print(f"Mediana da empresa 2: {mediana}")
print(f"Moda da empresa 2: {moda}")
print(f"Máximo da empresa 2: {max}")
print(f"Mínimo da empresa 2: {min}")
```

## Plotando o gráfico Bloxplot


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.boxplot(df)
plt.title('Boxplot das Empresas')
plt.ylabel('Valores')
plt.show()
```

```python
dft = df.transpose() # Invertendo linhas e colunas
plt.boxplot(dft)
plt.show()
```
